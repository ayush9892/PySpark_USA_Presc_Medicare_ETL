#### 1. Read us_cities_dimension dataframe

```
us_cities_dimension_df = spark.read \
    .parquet('/mnt/synapselearningadls/raw/pharmacy_usa/us_cities_dimension.parquet', inferSchema=True, header=True)

display(us_cities_dimension_df.limit(5))
```

#### 2. Select required column and convert into Upper Case

```
from pyspark.sql.functions import upper, col

us_cities_dimension_selected_df = us_cities_dimension_df \
    .select(upper(col('city')).alias('city'), upper(col('state_id')).alias('state_id'), 
            upper(col('state_name')).alias('state_name'), upper(col('county_name')).alias('county_name'), 'population', 'zips')
```

```
display(us_cities_dimension_selected_df.limit(5))
```

#### 3. Write processed data to processed container

```
us_cities_dimension_selected_df.write.mode('overwrite') \
    .format('delta').save('/mnt/synapselearningadls/processed/pharmacy_usa/us_cities_dimension')
```
